## Conditions

```python
qs = Shop.objects.all()
```

```python
# filter <-> exclude

qs.filter(pk=1)
```

```python
qs.exclude(pk=1)
```

```python
# lt : less than <
# lte : less than equal <=
# gt : greater than >
# gte : greater than equal >=

qs.filter(
    pk__gte=5, # greater than euqal
    pk__lt=10, # less than
)
```

```python
qs.filter(pk__gte=5).filter(pk__lt=10) # 체이닝
```

```python
qs1 = Shop.objects.all()
qs1 = qs1.filter(pk__gte=5)
qs1 = qs1.filter(pk__lt=10)
```

```python
qs2 = Shop.objects.all().filter(pk__gte=5).filter(pk__lt=10)
```

## Get 1 Row

## Conditions (OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10) # AND
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk__lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5) | Q(pk__lt=10)
)
```

```python
conditions = Q(pk__gte=5)
# conditions = conditions | Q(pk__lt=10)
conditions |= Q(pk__lt=10)

conditions
```

## 지정 범위의 Rows

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(
    description = "not used description",
)
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = "not used description"
    shop.save()
```
